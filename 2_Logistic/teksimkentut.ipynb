Metode Additive<br>
$Z_{i+1} = (Z_i + \mu) mod P$<br>
ada 2 parameter, $\mu$ dan P, tentuin sendiri $Z_0$, biar bagus, ambil $Z_0$ dan $\mu$ Prima, dan P prima yang besar

```python
Zo = 3
listOfRandomNumber = [Zo]
def additiveMethod(mu, p, length):
    for i in range(1, length):
        rand = (listOfRandomNumber[i-1] + mu) % p
        listOfRandomNumber.append(rand)
    return listOfRandomNumber
```

```python
t = additiveMethod(11, 1039, 1000)
t
```

Metode Inversi<br>
$x_i = \sqrt{2\cdot u_i}$

```python
import math
```

```python
def inversionMethod():
    inversionList = []
    for i in range(len(listOfRandomNumber)):
        inversionList.append(math.sqrt(2*listOfRandomNumber[i]))
    return inversionList
```
